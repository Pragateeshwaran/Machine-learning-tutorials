```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.linear_model import LogisticRegression
import pandas as pd 
from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
df.head()
```

```python
X, y = df.drop("DEATH_EVENT", axis=1), df[["DEATH_EVENT"]]
X_train, X_val, Y_train, Y_val = train_test_split(X, y, train_size=0.8, shuffle=False)
```

```python
SVC??
```

```python
model1 = GaussianNB()
model2 = LogisticRegression()
model3  = SVC(C=0.9, kernel="rbf")
model4 = RandomForestClassifier(n_estimators=10)
```

```python
model1.fit(X_train, Y_train)
model2.fit(X_train, Y_train)
model3.fit(X_train, Y_train)
model4.fit(X_train, Y_train)
```

```python
model1.score(X_train, Y_train), model2.score(X_train, Y_train), model3.score(X_train, Y_train), model4.score(X_train, Y_train)
```
